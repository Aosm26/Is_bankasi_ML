```python
from fastparquet import ParquetFile

# Train Parquet File
train_pf = ParquetFile("/kaggle/input/turkiye-is-bankasi-ml-challenge-5/train_final.parquet")
test_pf = ParquetFile("/kaggle/input/turkiye-is-bankasi-ml-challenge-5/test_final.parquet")
submission_pf = ParquetFile("/kaggle/input/turkiye-is-bankasi-ml-challenge-5/submission_sample_final.parquet")
# Convert the parquet files into dataframes
train_df = train_pf.to_pandas()
test_df = test_pf.to_pandas()
submission_df = submission_pf.to_pandas()
```
